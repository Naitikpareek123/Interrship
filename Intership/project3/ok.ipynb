```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\91785\Documents\GitHub\Naitikpareek123\python\project3\IMDB-Movie-Data.csv")
print(data)
```

```python
data.info()
```

```python
top_directors= data['Director'].value_counts().head(5)
print(top_directors)
```

```python
#13. Identify the 3 most frequent actors in the dataset.
actor_counts = data['Actors'].str.split(',').explode().value_counts().head(3)
print(actor_counts)
```

```python
# 14. Is there any visible relationship between box office earnings and movie ratings?
plt.scatter(data['Revenue (Millions)'], data['Rating'])
plt.xlabel('Revenue (Millions)')
plt.ylabel('Rating')
plt.show()
```

```python
# 20. Plot the average movie ratings over the years to identify trends.
avg_ratings_by_year = data.groupby('Year')['Rating'].mean()
avg_ratings_by_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```
